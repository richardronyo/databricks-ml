In this Notebook, I will be getting retrieving all of the rosters and team names for the WNBA 2026 season. I will be doing this by using the WeHoop package in R.

```python
#Getting the appropriate imports
from ryp import r
```

```python
# Setup R environment and install packages
r('''
    # 1. Set CRAN mirror
    options(repos = c(CRAN = "https://cloud.r-project.org/"))
    
    # 2. Define packages
    packages <- c("pak", "tictoc", "svglite")
    
    # 3. Install missing packages
    for (pkg in packages) {
        if (!require(pkg, character.only = TRUE)) {
            print(paste("Installing", pkg, "..."))
            install.packages(pkg, dependencies = TRUE)
        } else {
            print(paste(pkg, "is already installed"))
        }
    }
    
    # 4. Load all packages
    lapply(packages, library, character.only = TRUE)
    print("All packages ready!")
''')

# Now your Python code can use the R packages
print("R packages are ready for use!")
```

```python
#Must use pak to load wehoop from GitHub. Version 2.1.0 is on CRAN, but it does not contain the functions to load the roster and team data
r('''
    pak::pak("sportsdataverse/wehoop")
''')
```

```python
#Retrieving the Rosters and Teams data
r('''
    teams <- wehoop::fox_wnba_teams()
''')
```

```python
#Writing the Team data to a csv
r('''
    write.csv(teams, "data/wnba_teams.csv", row.names = FALSE)
''')
```

I will retrieve the roster for each team and save it to a csv for later processing

```python
r('''
    for (i in 1:30)
        {
            roster <- wehoop::fox_wnba_team_roster(as.character(i))
            write.csv(roster, paste0("data/rosters/team", i, "_roster.csv"), row.names = FALSE)
        }
''')
```
