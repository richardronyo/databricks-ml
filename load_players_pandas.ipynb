```python
teams = pd.read_csv("data/wnba_teams.csv")
teams = teams.drop(columns = "fox_section")
teams = teams.rename(columns = {"fox_team_id": "team_id",
                      "fox_team_name": "team_name"})
teams.head(10)
```

```python
import os

roster_dfs = []

for roster_file in os.listdir("data/rosters"):
    roster_df = pd.read_csv(f"data/rosters/{roster_file}")
    roster_dfs.append(roster_df)

rosters = pd.concat(roster_dfs, ignore_index = True)
rosters.head(20)
```

```python
complete_player_list = pd.merge(rosters, teams, how = "left", on = "team_id")
complete_player_list.head(50)
```
